```python
# // find the factorial of given number
```

```python
def facto(n):
    multi=1;
    while n>0:
        multi=multi*n;
        n=n-1;
    return multi;
print(facto(5));    
        
    
```

```python
# check prime or not
```

```python
def prime(n):
    if n<=1:
        return False;
    c=2    
    while c*c<=n:
        if n%c==0:
            return False;
        else:
            c=c+1;
    return True;

print(prime(13));    
```

```python
num=15
if num<=1:
    print('not prime')
else:
    isprime=True;
    for i in range(2,int(num**0.5)+1):
        if num%i==0:
            isprime=False
    if isprime:
        print('is prime')
    else:
        print('not prime')
        
```
