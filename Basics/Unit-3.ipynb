# QUE-1)input 10 numbers using arbitary arguments and find out occurance of perticular element in occarance

```python

def num(*args):
    mini=args[0];
    maxi=args[0];
    for i in range(0,len(args)):
        if args[i]<mini:
            mini=args[i];
        if args[i]>maxi:
            maxi=args[i];
    print(mini)
    print(maxi)
        
num(3,4,5,6,7,3,1,10,11,12);        
    
```

# QUE-4 FIND OUT THE LARGEST NUMBER BETWEEN TWO NUMBER USING LAMBDA FUNCTION

```python
a=lambda x,y:"x is gretaer" if x>y else  "y is greater";
print(a(2,4))
```

```python
a=lambda x,y,z:x if (x>y and x>z) else (y if y>z and y>x else z);
print(a(13,5,4))
```
